# Conhecendo a base de dados

## Importando os dados


```python
import pandas as pd
```

O método `read_csv` lê dados de um arquivo CSV e os converte em um *DataFrame*, uma estrutura de dados tabular bidimensional.

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
pd.read_csv(url, delimiter=',')
```

```python
pd.read_csv(url, delimiter=';')
```

```python
dados = pd.read_csv(url, delimiter=';')
dados
```

O método `head` é usado para exibir as primeiras linhas de um DataFrame. Por padrão, ele retorna as **primeiras cinco linhas**, mas podemos especificar um número diferente de linhas a serem exibidas, passando o número desejado como argumento para a função.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html

```python
dados.head()
```

```python
dados.head(10)
```

O método `tail` é usado para exibir as últimas linhas de um *DataFrame*. Similar ao `head`, por padrão, ele retorna as **últimas cinco linhas**, mas podemos especificar um número diferente de linhas a serem exibidas, passando o número desejado como argumento para a função.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.tail.html

```python
dados.tail()
```

O *DataFrame* é uma tabela de dados bidimensional, semelhante a uma planilha do Excel, com colunas rotuladas e linhas numeradas.

```python
type(dados)
```

## Características gerais da base de dados

O atributo `shape` é utilizado para obter as dimensões (número de linhas e colunas) de um *DataFrame*. O retorno será uma tupla no formato `(número de linhas, número de colunas)`

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shape.html

```python
dados.shape
```

O atributo `columns` é utilizado para acessar os rótulos (nomes) das colunas de um *DataFrame*.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.columns.html

```python
dados.columns
```

O método `info` é usado para obter informações sobre um *DataFrame*, incluindo os tipos de dados de cada coluna, a quantidade de valores nulos e o uso de memória.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html

```python
dados.info()
```

### Analisando colunas específicas

Para selecionar uma ou várias colunas de um DataFrame, podemos utilizar a notação de **colchetes** (`[]`).

* Colchetes simples (`[]`): apresentação de uma coluna;
* Colchetes 'duplos' (`[[]]`): usado para determinar mais de uma coluna, assemelhando-se às **listas de listas**.

```python
dados['Tipo']
```

```python
dados[['Quartos', 'Valor']]
```
