# Conhecendo a base de dados

## Importando os dados


```python
import pandas as pd
```

O método `read_csv` lê dados de um arquivo CSV e os converte em um *DataFrame*, uma estrutura de dados tabular bidimensional.

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
pd.read_csv(url, delimiter=',')
```

```python
pd.read_csv(url, delimiter=';')
```

```python
dados = pd.read_csv(url, delimiter=';')
dados
```

O método `head` é usado para exibir as primeiras linhas de um DataFrame. Por padrão, ele retorna as **primeiras cinco linhas**, mas podemos especificar um número diferente de linhas a serem exibidas, passando o número desejado como argumento para a função.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html

```python
dados.head()
```

```python
dados.head(10)
```

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.tail.html

```python
dados.tail()
```

O *DataFrame* é uma tabela de dados bidimensional, semelhante a uma planilha do Excel, com colunas rotuladas e linhas numeradas.

# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

O método `mean` é usado para calcular a média artimética ao longo de um ou mais eixos específicos de um *DataFrame*.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mean.html

```python
dados['Valor'].mean()
```

O método `groupby` permite agrupar e resumir dados de um *DataFrame*, com base em um ou mais critérios — que podem ser, por exemplo, variáveis numéricas ou categóricas, como colunas ou índices do *DataFrame*.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
dados.groupby('Tipo')['Valor'].mean()
```

O método `sort_values` é utilizado para ordenar um *DataFrame* com base nos valores de uma ou mais colunas.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='#9CD33A')
```

## Removendo os imóveis comerciais

O método `unique` é utilizado para obter os **valores únicos** contidos em uma coluna específica de um *DataFrame*. O retorno é um array.

https://pandas.pydata.org/docs/reference/api/pandas.unique.html

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

O método `query` é utilizado para filtrar linhas de um *DataFrame* que atendem a uma condição específica.

> **Observação:** A utilização do `@` antes do nome da variável indica que você está se referindo a uma variável do ambiente Python dentro da expressão de consulta.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html

```python
dados.query('@imoveis_comerciais in Tipo')
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='#9CD33A')
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df.Tipo.unique()
```

O método `value_counts` é usado para contar quantas vezes determinado valor aparece e uma coluna específica de um *DataFrame*. O retorno é em formato de *Series*.

https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html

```python
df.Tipo.value_counts(normalize=True)
```

O método `to_frame` é usado para converter uma *Seris* em um *DataFrame*.

https://pandas.pydata.org/docs/reference/api/pandas.Series.to_frame.html

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipo.plot(kind='bar', figsize=(14, 10), color='#9CD33A', xlabel='Tipos', ylabel='Percentual')
```
